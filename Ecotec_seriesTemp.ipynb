```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# URL del archivo Excel
url = "https://github.com/santiagonajera/AnalisisSeriesTemporales/raw/refs/heads/main/ventas_diarias_2021_2022.xlsx"

# Cargar el archivo Excel
df = pd.read_excel(url)

# Convertir la columna 'Fecha' a datetime
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Extraer año y mes de la columna 'Fecha'
df['Año'] = df['Fecha'].dt.year
df['Mes'] = df['Fecha'].dt.month

# Agrupar los datos por año y mes para todos los productos
productos = [col for col in df.columns if col.startswith('Producto')]
ventas_mensuales = df.groupby(['Año', 'Mes'])[productos].sum().reset_index()

# Crear una columna combinada de fecha (año-mes)
ventas_mensuales['Fecha_Mensual'] = pd.to_datetime(
    ventas_mensuales['Año'].astype(str) + '-' + ventas_mensuales['Mes'].astype(str) + '-01'
)

# Ordenar los datos por fecha
ventas_mensuales = ventas_mensuales.sort_values(by='Fecha_Mensual')

# Preparar los datos para la regresión lineal
ventas_mensuales['Ordinal'] = ventas_mensuales['Fecha_Mensual'].apply(lambda x: x.toordinal())

# Función para calcular el forecast de un producto
def calcular_forecast(producto):
    # Variables independiente (X) y dependiente (y)
    X = ventas_mensuales[['Ordinal']]
    y = ventas_mensuales[producto]

    # Crear y ajustar el modelo de regresión lineal para la tendencia
    modelo_tendencia = LinearRegression()
    modelo_tendencia.fit(X, y)

    # Predecir los valores de la tendencia
    tendencia = modelo_tendencia.predict(X)

    # Calcular la estacionalidad multiplicativa
    estacionalidad = y / tendencia

    # Promediar la estacionalidad por mes para capturar el patrón estacional
    estacionalidad_promedio = ventas_mensuales.groupby('Mes')[producto].transform(lambda x: x.mean())

    # Calcular el forecast para los datos existentes
    forecast_existente = tendencia * estacionalidad_promedio

    return tendencia, estacionalidad_promedio, forecast_existente

# Crear un DataFrame para almacenar las proyecciones de los próximos 6 meses
proyecciones = []

# Calcular el forecast para cada producto
for producto in productos:
    # Calcular tendencia, estacionalidad y forecast existente
    tendencia, estacionalidad_promedio, _ = calcular_forecast(producto)

    # Obtener la última fecha mensual
    ultima_fecha = ventas_mensuales['Fecha_Mensual'].iloc[-1]

    # Generar las próximas 6 fechas
    proximas_fechas = pd.date_range(start=ultima_fecha + pd.offsets.MonthBegin(), periods=6, freq='MS')
    proximos_ordinales = proximas_fechas.to_series().apply(lambda x: x.toordinal())

    # Calcular la tendencia para las próximas fechas
    proxima_tendencia = modelo_tendencia.predict(proximos_ordinales.values.reshape(-1, 1))

    # Obtener la estacionalidad promedio para los próximos meses
    proxima_estacionalidad = [
        estacionalidad_promedio.iloc[(mes - 1) % 12] for mes in proximas_fechas.month
    ]

    # Calcular el forecast para los próximos 6 meses
    proximo_forecast = proxima_tendencia * proxima_estacionalidad

    # Guardar las proyecciones en una lista
    for fecha, valor in zip(proximas_fechas, proximo_forecast):
        proyecciones.append({'Producto': producto, 'Fecha': fecha, 'Forecast': valor})

# Crear un DataFrame con las proyecciones
proyecciones_df = pd.DataFrame(proyecciones)

# Mostrar las proyecciones de los próximos 6 meses para todos los productos
print(proyecciones_df.pivot(index='Fecha', columns='Producto', values='Forecast'))
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# URL del archivo Excel
url = "https://github.com/santiagonajera/AnalisisSeriesTemporales/raw/refs/heads/main/ventas_diarias_2021_2022.xlsx"

# Cargar el archivo Excel
df = pd.read_excel(url)

# Convertir la columna 'Fecha' a datetime
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Extraer año y mes de la columna 'Fecha'
df['Año'] = df['Fecha'].dt.year
df['Mes'] = df['Fecha'].dt.month

# Agrupar los datos por año y mes para todos los productos
productos = [col for col in df.columns if col.startswith('Producto')]
ventas_mensuales = df.groupby(['Año', 'Mes'])[productos].sum().reset_index()

# Crear una columna combinada de fecha (año-mes)
ventas_mensuales['Fecha_Mensual'] = pd.to_datetime(
    ventas_mensuales['Año'].astype(str) + '-' + ventas_mensuales['Mes'].astype(str) + '-01'
)

# Ordenar los datos por fecha
ventas_mensuales = ventas_mensuales.sort_values(by='Fecha_Mensual')

# Preparar los datos para la regresión lineal
ventas_mensuales['Ordinal'] = ventas_mensuales['Fecha_Mensual'].apply(lambda x: x.toordinal())

# Función para calcular el forecast de un producto
def calcular_forecast(producto):
    # Variables independiente (X) y dependiente (y)
    X = ventas_mensuales[['Ordinal']].values  # Usamos .values para evitar advertencias
    y = ventas_mensuales[producto].values     # Usamos .values para evitar advertencias

    # Crear y ajustar el modelo de regresión lineal para la tendencia
    modelo_tendencia = LinearRegression()
    modelo_tendencia.fit(X, y)

    # Predecir los valores de la tendencia
    tendencia = modelo_tendencia.predict(X)

    # Calcular la estacionalidad multiplicativa
    estacionalidad = y / tendencia

    # Promediar la estacionalidad por mes para capturar el patrón estacional
    estacionalidad_promedio = ventas_mensuales.groupby('Mes')[producto].transform('mean') / tendencia

    # Obtener la última fecha mensual
    ultima_fecha = ventas_mensuales['Fecha_Mensual'].iloc[-1]

    # Generar las próximas 6 fechas
    proximas_fechas = pd.date_range(start=ultima_fecha + pd.offsets.MonthBegin(), periods=6, freq='MS')
    proximos_ordinales = proximas_fechas.to_series().apply(lambda x: x.toordinal()).values.reshape(-1, 1)

    # Calcular la tendencia para las próximas fechas
    proxima_tendencia = modelo_tendencia.predict(proximos_ordinales)

    # Obtener la estacionalidad promedio para los próximos meses
    proxima_estacionalidad = [
        estacionalidad_promedio.iloc[(mes - 1) % 12] for mes in proximas_fechas.month
    ]

    # Calcular el forecast para los próximos 6 meses
    proximo_forecast = proxima_tendencia * proxima_estacionalidad

    return proximas_fechas, proximo_forecast

# Crear un DataFrame para almacenar las proyecciones de los próximos 6 meses
proyecciones = []

# Calcular el forecast para cada producto
for producto in productos:
    proximas_fechas, proximo_forecast = calcular_forecast(producto)
    for fecha, valor in zip(proximas_fechas, proximo_forecast):
        proyecciones.append({'Producto': producto, 'Fecha': fecha, 'Forecast': valor})

# Crear un DataFrame con las proyecciones
proyecciones_df = pd.DataFrame(proyecciones)

# Formatear los valores de Forecast con separadores de miles
proyecciones_df['Forecast'] = proyecciones_df['Forecast'].map('{:,.2f}'.format)

# Mostrar las proyecciones de los próximos 6 meses para todos los productos
print(proyecciones_df.pivot(index='Fecha', columns='Producto', values='Forecast'))
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from tabulate import tabulate

# Configuración de visualización
pd.set_option('display.float_format', '{:,.2f}'.format)

# URL del archivo Excel
url = "https://github.com/santiagonajera/AnalisisSeriesTemporales/raw/refs/heads/main/ventas_diarias_2021_2022.xlsx"

# Cargar y preparar los datos
def cargar_datos(url):
    df = pd.read_excel(url)
    df['Fecha'] = pd.to_datetime(df['Fecha'])
    df['Año'] = df['Fecha'].dt.year
    df['Mes'] = df['Fecha'].dt.month
    return df

# Procesar datos mensuales
def procesar_datos_mensuales(df):
    productos = [col for col in df.columns if col.startswith('Producto')]
    ventas_mensuales = df.groupby(['Año', 'Mes'])[productos].sum().reset_index()
    ventas_mensuales['Fecha_Mensual'] = pd.to_datetime(
        ventas_mensuales['Año'].astype(str) + '-' + ventas_mensuales['Mes'].astype(str) + '-01'
    )
    ventas_mensuales = ventas_mensuales.sort_values('Fecha_Mensual')
    ventas_mensuales['Ordinal'] = ventas_mensuales['Fecha_Mensual'].apply(lambda x: x.toordinal())
    return ventas_mensuales, productos

# Modelo de pronóstico
def generar_pronosticos(ventas_mensuales, productos, meses_pronostico=6):
    proyecciones = []

    for producto in productos:
        X = ventas_mensuales[['Ordinal']].values
        y = ventas_mensuales[producto].values

        modelo = LinearRegression()
        modelo.fit(X, y)
        tendencia = modelo.predict(X)
        estacionalidad = y / tendencia

        # Estacionalidad promedio por mes
        ventas_mensuales['Estacionalidad'] = estacionalidad
        estacionalidad_promedio = ventas_mensuales.groupby('Mes')['Estacionalidad'].mean()

        # Generar fechas futuras
        ultima_fecha = ventas_mensuales['Fecha_Mensual'].iloc[-1]
        proximas_fechas = pd.date_range(
            start=ultima_fecha + pd.offsets.MonthBegin(),
            periods=meses_pronostico,
            freq='MS'
        )

        # Calcular pronósticos
        proximos_ordinales = np.array([f.toordinal() for f in proximas_fechas]).reshape(-1, 1)
        proxima_tendencia = modelo.predict(proximos_ordinales)
        meses_futuros = [f.month for f in proximas_fechas]
        proxima_estacionalidad = estacionalidad_promedio[meses_futuros].values

        pronosticos = proxima_tendencia * proxima_estacionalidad

        for fecha, pronostico in zip(proximas_fechas, pronosticos):
            proyecciones.append({
                'Producto': producto,
                'Fecha': fecha.strftime('%Y-%m-%d'),
                'Pronóstico': pronostico
            })

    return pd.DataFrame(proyecciones)

# Presentación de resultados
def presentar_resultados(proyecciones_df):
    pivot_df = proyecciones_df.pivot(index='Fecha', columns='Producto', values='Pronóstico')

    # Formatear como tabla con bordes
    print("\nPRONÓSTICO DE VENTAS - PRÓXIMOS 6 MESES\n")
    print(tabulate(pivot_df, headers='keys', tablefmt='grid', floatfmt=",.2f"))

    # Opcional: Exportar a Excel
    pivot_df.to_excel("pronosticos_ventas.xlsx")
    print("\nLos resultados se han exportado a 'pronosticos_ventas.xlsx'")

# Flujo principal
def main():
    print("Cargando y procesando datos...")
    df = cargar_datos(url)
    ventas_mensuales, productos = procesar_datos_mensuales(df)

    print("Generando pronósticos...")
    proyecciones_df = generar_pronosticos(ventas_mensuales, productos)

    presentar_resultados(proyecciones_df)

if __name__ == "__main__":
    main()
```
